```python
import pandas as pd

housing_1 = pd.read_csv('housing_1.csv',sep='\t')
housing_1
```

```python
address1 = pd.read_csv('address1.csv',sep="\t")
```

```python
df = pd.concat([housing_1,address1[["Latitude","Longitude"]]],axis=1)
```

```python
df.rename(columns={"Latitude":"lat1","Longitude":"lon1"},inplace=True)
```

```python
df.to_csv('housing_1.csv',sep='\t',index=False)
```
